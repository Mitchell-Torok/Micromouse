### Task 6 - Dilate

Perform an dilate operation from the source image. Use a kernel size of 3 and perform 5 iterations. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

IMAGE_FILE = "maze.jpg"
thresh = 128
maxval = 255

img_maze = cv2.imread(IMAGE_FILE)
img_hsv = cv2.cvtColor(img_maze, cv2.COLOR_BGR2HSV)

hsv_mask = cv2.inRange(img_hsv, (0,4,0), (23,255, 255))

_, threshed = cv2.threshold(hsv_mask, thresh, maxval, cv2.THRESH_BINARY)
threshed_3_channel = cv2.cvtColor(threshed, cv2.COLOR_GRAY2BGR)

plt.figure(figsize = (9,5))
plt.imshow(threshed_3_channel)
```

```python
img = cv2.cvtColor(threshed, cv2.COLOR_GRAY2BGR)
kernel = np.ones((3,3), np.uint8)

imgNew = cv2.dilate(img, kernel, iterations = 1)

kernel_size = 3
kernel = np.ones((kernel_size, kernel_size), np.uint8)
imgNew = cv2.erode(img, kernel, iterations = 2)

kernel_size = 7

kernel = np.ones((kernel_size, kernel_size), np.uint8)
img = cv2.dilate(img, kernel, iterations = 1)

kernel_size = 6
kernel = np.ones((kernel_size, kernel_size), np.uint8)
img = cv2.erode(img, kernel, iterations = 8)

fig, (ax1, ax2) = plt.subplots(figsize = (9, 5), ncols = 2)
ax1.imshow(img, cmap = 'gray'), ax1.set_title("Original")
ax2.imshow(img, cmap='gray'), ax2.set_title("Erosion")

```
